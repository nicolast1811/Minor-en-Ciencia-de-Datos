Con while (iterativo)




```python
fib0 = 0
fib1 = 1
n = int(input('Ingrese n = '))
print('Secuencia de Fibonacci para los primeros', n, 'números: ')
i = 0
print('Para el valor', i,'-->', fib0) #valor de 0 es siempre el mismo
while i < n:  
  fib_n = fib1 + fib0
    #Se aumentan los valores para que hagamos siempre la misma operación pero con
    #los valores actualizados
  fib0 = fib1
  fib1 = fib_n
  print('Para el valor', i+1,'-->', fib0)
  i+=1
```

Con función (recursivo)

```python
def fibo(n):
  if n <= 1:
    return n
  else:
    return (fibo(n-2) + fibo(n-1))

num = int(input('Ingrese valor = '))
for i in range(num+1):
  print('Para el valor', i,'-->', fibo(i))
```

```python
#one liner (GPT)

fibonacci = lambda n: n if n <= 1 else fibonacci(n-1) + fibonacci(n-2)

for n in range(16):
    print(fibonacci(n))
```
