```python
# Generar un KDE plot para la variable 'edad' en df
sns.kdeplot(df['edad'], shade=True)
plt.title("Distribución KDE de la variable 'edad'")
plt.xlabel("Edad")
plt.ylabel("Densidad")
plt.show()
import scipy.stats as stats
data_column = df['edad']  # Ajusta el nombre de la columna
# Realiza el test de Anderson-Darling
result = stats.anderson(data_column, dist='norm')

# Imprime el resultado
print("Estadístico de Anderson-Darling:", result.statistic)
print("Valores críticos:", result.critical_values)
print("Niveles de significancia:", result.significance_level)
```

Con estos resultados, al ser cada valor crítico menor al Estadístico, se observa que edad no sigue una distribución normal

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Seleccionar las variables para realizar QQ plots: 'habituales', 'habituales', 'edad'
selected_columns = ['efectivas', 'habituales', 'edad']

# Generar QQ plots, realizar pruebas de normalidad y de asimetría para cada variable seleccionada
for col in selected_columns:
    data = df_codi[col].dropna()  # Eliminar valores nulos

    # Realizar una prueba de normalidad adecuada
    if len(data) < 5000:
        # Prueba de Shapiro-Wilk para muestras más pequeñas
        stat, p_value = stats.shapiro(data)
        test_name = "Shapiro-Wilk"
    else:
        # Prueba de Kolmogorov-Smirnov para muestras grandes
        stat, p_value = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
        test_name = "Kolmogorov-Smirnov"

    # Mostrar resultados de la prueba de normalidad
    print(f"Prueba de normalidad {test_name} para {col}:")
    print("Estadístico:", stat)
    print("p-valor:", p_value)
    if p_value > 0.05:
        print(f"La variable {col} podría seguir una distribución normal.")
    else:
        print(f"La variable {col} no sigue una distribución normal.")
    print("\n")

    # Prueba de asimetría (skew)
    skewness = stats.skew(data)
    print(f"Asimetría (skew) para {col}: {skewness}")
    if skewness > 0:
        print(f"La distribución de {col} está sesgada positivamente.")
    elif skewness < 0:
        print(f"La distribución de {col} está sesgada negativamente.")
    else:
        print(f"La distribución de {col} es simétrica.")
    print("\n")

    # Generar el QQ plot
    plt.figure()
    stats.probplot(data, dist="norm", plot=plt)
    plt.title(f"QQ Plot de {col}")
    plt.show()
```

Un sesgo negativo indica que la variable tiene algunos valores excepcionalmente bajos que afectan la simetría de la distribución, pero que la mayoría de los datos están concentrados en el rango de valores altos.

### **MCA**

```python
import prince
# Convertir las variables categóricas a variables dummies
#df_dummies = pd.get_dummies(df_filtrado_imputed, dummy_na=True)
df_dummies = pd.get_dummies(df_filtrado)
#MCA
mca = prince.MCA(n_components=2, n_iter=3, copy=True, check_input=True, engine='sklearn', random_state=42)
mca = mca.fit(df_dummies)
# Transformar los datos para obtener el conjunto reducido
df_mca = mca.transform(df_dummies)
# Mostrar los primeros resultados de los componentes principales obtenidos por MCA
print(df_mca.head())
```

```python
import pandas as pd
import prince
# Obtener los eigenvalues (valores propios)
eigenvalues = mca.eigenvalues_
print("Eigenvalues (valores propios) de cada componente:", eigenvalues)

# Calcular el porcentaje de inercia explicada por cada componente
total_inertia = sum(eigenvalues)
explained_inertia_percentage = [(eigenvalue / total_inertia) * 100 for eigenvalue in eigenvalues]

# Mostrar los porcentajes de inercia explicada
print("Porcentaje de inercia explicada por cada componente:", explained_inertia_percentage)
```
