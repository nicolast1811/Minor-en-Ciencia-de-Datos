```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris)
type(iris)
```

```python
x = iris.data 
y = iris.target 

print(x[:6])

print(y[:6])
```

```python
import numpy as np
data = np.column_stack((x,y))
print(data)
```

```python
import pandas as pd
df = pd.DataFrame(data)

print(df)
```
