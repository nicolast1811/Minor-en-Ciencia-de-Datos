```python
# a = int(input('ingrese valor a = '))
# b = int(input('ingrese valor b = '))
# c = int(input('ingrese valor c = '))
def whilec2(a,b,c):
  contador = 0
  i=0
  while i < c:
    contador += a*b
    i+=1
  return contador

whilec2(6.3,1,2)
```
